Tensor operations include
* addition
* subtraction
* multiplication (elemnet-wise)
* division
* matrix multiplication

There are two main rules mat mul should satisfy:
1. Inner dimensions must match
* `(3,2) @ (3,2) will not work`
* `(2,3) @ (3,2) will work`
* `(3,2) @ (2,3) will work`
2. The resulting matrix has shape of outer dimensions

### Reshaping, Viewing and Stacking, Squeezing and unsqueezing

* Reshaping -> reshapes an input tensor to defined shape
* View -> Return a view of an input tensor of certain shape but keep the same memory as original tensor
* Stacking -> Combine multiple tensors on top of each other (vstack), or side by side (hstack)
* Squeeze -> removes all `1` dimensions from a tensor
* Unsqueezze -> add a `1` dimension to a target tensor
* Permute -> Return a View of the input with dimensions permuted (swapped) in a certain way

```python
x = torch.arange(1.,10.)
x,x.shape
```

```python
#new dimensions have to be compatible with original dimensions.
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

```python
#Change the view
z=x.view(1,9)
z, z.shape
```

```python
#changing z changes x because a view of tensor shares the same memory as original

z[:,0]=5
z,x
```

```python
#stack tensors on top of each other
x_stacked= torch.stack([x,x,x,x], dim=0) #vertical
x_stacked
```

```python
x_stacked= torch.stack([x,x,x,x], dim=1) #horizontal #dim2 incompatible bcz x only has 0,1 dim
x_stacked
```

```python
#torch.squeeze removes all single dimensions from target tensor
x_reshaped, x_reshaped.shape
```

```python
x_reshaped.squeeze(), x_reshaped.squeeze().shape
```

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous tensor shape: {x_reshaped.shape}")
x_squeezed = x_reshaped.squeeze()
print(f"Squeeze tensor: {x_squeezed}")
print(f"Squeeze tensor shape: {x_squeezed.shape}")
```

```python
#torch.qunsqueeze adds a single dimension to a target tensor at a specific dimension
print(f"previous tensor: {x_squeezed}")
print(f"previous shape: {x_squeezed.shape}")
x_unsqueezed= x_squeezed.unsqueeze(dim=1)
print(f"New tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
x_unsqueezed= x_squeezed.unsqueeze(dim=0)
print(f"New tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")

```

```python
#torch.permute rearranges the dimensions of a target tensor in a specified order
#used very much with images
x_orig = torch.rand(size=(224,224, 3)) # heught, widyh, color channels
#Permute the orig to rearrange the axis(or dim) order
x_permuted = x_orig.permute(2,0,1) #shifts 0->1, 1->2, 2->0

"""since permute is a view, they share memory, if you chnage an elemnt in one,
it changes in both"""
x_orig.shape, x_permuted.shape
```

### Pytorch Reproducibility (Tying to take the random out of random)

How a NN learns:
`start with random numbers-> tensor operations -> update random numbers to try and make them a better representation of data -> again -> again -> again...`

How to make the results reproducible EXACTLY?

To reduce the randomnes in NN and PyTorch, comes the concept of a *random seed*. Essentially what random seed does is "flavour" the randomness.

```python
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)
random_tensor_A, random_tensor_B, random_tensor_A==random_tensor_B
```

```python
# Let's make some random but reproducible tensors

RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)

random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED) #needed everytime!!!!!
random_tensor_D = torch.rand(3,4)
random_tensor_C, random_tensor_D, random_tensor_C==random_tensor_D
```

```python
torch.rand(3,4)
```

```python
torch.rand(3,4)
```

```python
torch.manual_seed(RANDOM_SEED)
torch.rand(3,4)
```
